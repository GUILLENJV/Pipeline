No tenemos ningún campo que identifique exclusivamente la propiedad. Dado que nuestro modelo no se beneficiará de los valores duplicados, vamos a eliminarlos y mantener el primer valor.

```python
df = df.drop_duplicates(keep='first')
```

```python
df.shape
```

## Validación cruzada de campos
Dado que nuestro objetivo es el precio del alquiler de la propiedad, vamos a comprobar la integridad de las columnas monetarias relacionadas con el importe del alquiler.

Normalmente, estos valores no pueden ser más caros que el precio del alquiler, así que vamos a comprobarlo.

```python
df.loc[df['property tax (R$)'] > df['rent amount (R$)']]
```

```python
df.loc[df['fire insurance (R$)'] > df['rent amount (R$)']]
```

```python
df.loc[df['hoa (R$)'] > df['rent amount (R$)']]
```

Podemos ver que sólo la columna hoa tiene valores más caros que el precio del alquiler.

Para confirmar este comportamiento, investigamos un poco y descubrimos que es posible que el precio del alquiler sea más caro que el del alquiler, por lo que en este caso no eliminaremos estos valores.
